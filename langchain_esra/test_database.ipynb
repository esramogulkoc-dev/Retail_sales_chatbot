## Database Tests — Simple SQL Queries
Testing the `data/tshirts.db` SQLite database.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('data/tshirts.db')
print('Connected to tshirts.db')
```

```python
# All t-shirts
pd.read_sql_query('SELECT * FROM t_shirts LIMIT 10', conn)
```

```python
# All discounts
pd.read_sql_query('SELECT * FROM discounts', conn)
```

```python
# Total stock per brand
pd.read_sql_query('''
    SELECT brand, SUM(stock_quantity) AS total_stock
    FROM t_shirts
    GROUP BY brand
    ORDER BY total_stock DESC
''', conn)
```

```python
pd.read_sql_query('''
    SELECT SUM(price*stock_quantity)
    FROM t_shirts
    WHERE size = 'S'
''', conn)
```

```python
pd.read_sql_query('''
    SELECT SUM(price*stock_quantity)
    FROM t_shirts
    WHERE size = 'XL'
''', conn)
```

```python
# Total stock per size
pd.read_sql_query('''
    SELECT size, SUM(stock_quantity) AS total_stock
    FROM t_shirts
    GROUP BY size
    ORDER BY total_stock DESC
''', conn)
```

```python
# Total inventory value (price x stock)
pd.read_sql_query('''
    SELECT brand, SUM(price * stock_quantity) AS inventory_value
    FROM t_shirts
    GROUP BY brand
    ORDER BY inventory_value DESC
''', conn)
```

```python
# T-shirts with discounts (JOIN)
pd.read_sql_query('''
    SELECT t.brand, t.color, t.size, t.price, d.pct_discount,
           ROUND(t.price * (1 - d.pct_discount / 100), 2) AS discounted_price
    FROM t_shirts t
    JOIN discounts d ON t.t_shirt_id = d.t_shirt_id
''', conn)
```

"How much sales will we generate if we sell all Adidas shirts?"

```python
pd.read_sql_query('''
SELECT SUM(t.price * (1 - d.pct_discount / 100) * t.stock_quantity) AS total_discounted_revenue
FROM t_shirts t
JOIN discounts d ON t.t_shirt_id = d.t_shirt_id
WHERE t.brand = 'Van Huesen';
''', conn)
```

```python
pd.read_sql_query('''
SELECT SUM(t.price * (1 - d.pct_discount / 100) * t.stock_quantity) AS total_discounted_revenue
FROM t_shirts t
JOIN discounts d ON t.t_shirt_id = d.t_shirt_id
WHERE t.brand = 'Levi';
''', conn)
```
